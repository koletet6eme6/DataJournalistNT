```python
logs.traffic_source.unique()
```

```python
def trafic_type(element):
    if (element['traffic_source'] == 'yandex') or (element['traffic_source'] == 'google'):
        return 'organic'
    elif (element['traffic_source'] == 'paid') or (element['traffic_source'] == 'email'):
        if (element ['region'] == 'Russia'):
            return 'ad'
    elif (element['traffic_source'] == 'paid') or (element['traffic_source'] == 'mail'):
        if (element['region'] != 'Russia'):
            return 'other'
    else:
        return element['traffic_source']
    
        
    
```

```python
logs['source_type'] = logs.apply(trafic_type, axis = 1)
logs.head()
```
