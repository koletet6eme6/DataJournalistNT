```python
res = requests.get('https://habr.com/ru/all/')
res
```

```python
res.text
```

```python
soup = BeautifulSoup(res.text, 'html.parser')
```

```python
pb = soup.find_all('article', class_ = 'post post_preview')
len(pb)
```

```python
#получаем блоки с текстами превью
posts_blocks = soup.find_all('div', class_ = 'post__body post__body_crop')
len(posts_blocks)
```

```python
#получаем тексты превью
preview_texts = []
for element in posts_blocks:
    preview_texts.append(element.get_text())
preview_texts
```

```python
#получаем блоки с ссылками (ищем два класса, так как один блок иначе теряется)
links_blocks = soup.find_all('h2', {'class':['post__title', 'preview-data__title']})
len(links_blocks)
```

```python
#получаем ссылки
links_list = list(map(lambda x: x.find('a').get('href'), links_blocks))
len(links_list)
```

```python
# получаем заголовки
headers = []
for element in links_blocks:
    element.find(class_ = 'post__title_link') and element.find(class_ = 'preview-data__title')
    element = element.get_text().strip()
    headers.append(element)
len(headers)
```
