```python
#задача 2
data = pd.read_csv('power.csv')
```

```python
filtered_countries = data [ (data['country'] == 'Latvia') | (data['country'] == 'Lithuania') | (data['country'] == 'Estonia')]
filtered_category = filtered_countries[ (filtered_countries['category'] == 4) | (filtered_countries['category'] == 12) | (filtered_countries['category'] == 21)]
filtered_quantity = filtered_category [filtered_category['quantity'] > 0]
filtered_by_year = filtered_quantity[(filtered_quantity['year'] >= 2005) & (filtered_quantity['year'] <= 2010)]
#сумма по всем странам Прибалтики
quantity_sum = filtered_by_year['quantity'].sum()                        
print(f'Суммарное потребление стран Прибалтики категорий 4, 12, 21 за период с 2005 по 2010 года составило: {quantity_sum}')
```
