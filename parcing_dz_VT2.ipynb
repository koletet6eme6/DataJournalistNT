```python
#находим посты
pb = soup.find_all('article', class_ = 'post post_preview')
pb
```

```python
#получаем тексты превью 
pd_blocks = list(map(lambda x: x.find('div', class_ = 'post__body post__body_crop').get_text(), pb))
len(pd_blocks)
```

```python
#получаем блоки с сылками
lnk_blocks = list(map(lambda x: x.find('h2', class_='post__title'), pb))
lnk_blocks
```

```python
#получаем ссылки
lnk_list = list(map(lambda x: x.find('a').get('href'), lnk_blocks))
lnk_list
```

```python
import pandas as pd
```

```python
my_df = pd.DataFrame()
for page in pages:
    # получаем страницу
    req = requests.get(page)
    soup = BeautifulSoup(req.text)
    # достаем посты
    pb = soup.find_all('article', class_='post post_preview')
    # находим посты с ключевыми словами
    for post in pb:
        for key in keywords:
            if key in post.text:
                # извлекаем дату, название статьи, ссылку и записываем в датафрейм
                date = post.find('span', class_='post__time').text
                if 'сегодня' in date:
                    date = date.replace('сегодня', str(datetime.date.today()))
                elif 'вчера' in date:
                    date = date.replace('вчера', str(datetime.date.today() - datetime.timedelta(days=1)))
                link = post.find('a', class_="post__title_link").get('href')
                title = post.find('a', class_= 'post__title_link').text
                row = {'date':date, 'title': title, 'link': link}
                my_df = pd.concat([my_df, pd.DataFrame([row])])
                
    
```
