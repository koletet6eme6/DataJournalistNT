```python
my_emails = {'emailAddresses': ['xxx@x.ru', 'yyy@y.com']}
```

```python
url = 'https://identityprotection.avast.com/v1/web/query/site-breaches/unauthorized-data'
```

```python
headers = {'Vaar-Version': '0', 'Vaar-Header-App-Product-Name': 'hackcheck-web-avast', 'Vaar-Header-App-Build-Version': '1.0.0'}
```

```python
avast_response = requests.post(url, json = my_emails, headers = headers)
```

```python
avast_response
```

```python
response = json.loads(avast_response.text)
response
```

```python
# делаем словарь, ключ которого адрес, а значение - список id утечек
```

```python
response['summary']
emails_and_breaches = {key:list(value.values())[0] for key, value in response['summary'].items()}
```

```python
emails_and_breaches
```

```python
my_df = pd.DataFrame()
```

```python
for mail, ids in emails_and_breaches.items():
    df_email = mail
    for breach, data in response['breaches'].items():
        if int(breach) in ids:
            data_df = data['publishDate']
            source_df = data['site']
            description_df = data['description']
            my_row = {'email':df_email, 'data': data_df, 'source': source_df, 'description': description_df}
            my_df = pd.concat([my_df, pd.DataFrame([my_row])])
```
