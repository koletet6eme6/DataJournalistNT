```python
# собираем ссылки на все материалы
def get_all_links(url):
    all_refs = []
    for i in range(79):
        i = i + 1
        res = requests.get(url)
        time.sleep(0.1)
        soup = BeautifulSoup(res.text, 'html.parser')
        links_blocks = soup.find_all('h2', class_ = 'post__title')
        links_list = list(map(lambda x: x.find('a').get('href'), links_blocks))
        all_refs += list(map(lambda x: x.find('a').get('href'), links_blocks))
    return all_refs

all_links = get_all_links(url)
```

```python
import pandas as pd
```

```python
#получаем дф (пока только с датами)
my_df = pd.DataFrame()
for link in all_links:
    soup = BeautifulSoup(requests.get(link).text, 'html.parser')
    time.sleep(0.1)
    date = pd.to_datetime(soup.find('span', class_ = 'post__time').get('data-time_published'), dayfirst = True).date()
    row = {'date':date}
    my_df = pd.concat([my_df, pd.DataFrame([row])]) 

```
